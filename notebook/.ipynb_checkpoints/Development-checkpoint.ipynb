```python
import numpy as np
import math
```

```python
class GPR:
    def __init__(self):
        self.x = None
        self.y = None
        
    def train(self,x,y):
        self.x,self.y = x,y
    
    def rbf_kernel(x1, x2, sigma=1):
        """An implementation of the radial basis function kernel.

        Args:
            a: A vector
            b: A vector
            sigma: The radius of the kernel
        """
        distance = np.dot(x1 - x2,x1 - x2)
        return math.exp(- distance / (2 * (sigma) ** 2))
    
    def m_kernel(self, X1, X2, kernel):
        n1,n2 = X1.shape[0],X2.shape[0]
        K = np.zeros((n1,n2))
        
        for i in range(n1):
            for j in range(n2):
                K[i,j] = kernel(X1[i],X2[j])
                
        return K
        
    def predict(self,x_test,kernel=rbf_kernel):
        A = self.m_kernel(self.x,self.x,kernel)
        Ainv = np.linalg.inv(A)
        B = self.m_kernel(x_test,self.x,kernel)
        C = self.m_kernel(x_test,x_test,kernel)
        
        mu_test = B@Ainv@self.y
        sigma_test = C - B@Ainv@B.T
        return mu_test,sigma_test
    
```

```python
import matplotlib.pyplot as plt
params = { 
    'axes.labelsize': "large",
    'xtick.labelsize': 'x-large',
    'legend.fontsize': 20,
    'figure.dpi': 150,
    'figure.figsize': [12, 4]
}
plt.rcParams.update(params)


reg = GPR()
x = np.linspace(0,2*math.pi,49).reshape(-1,1)

x_train,x_test = x[::2],x[1::2]
y_train,y_test = np.sin(x_train),np.sin(x_test)

reg.train(x_train,y_train)

y_pred,sigma_pred = reg.predict(x)

sigdiag = np.diag(sigma_pred)
plt.errorbar(x.reshape(-1),y_pred.reshape(-1),yerr=sigdiag,capsize=0)
plt.plot(x_train,y_train,"ro")
plt.show()
```
